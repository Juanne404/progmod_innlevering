## lesing og plotting av data

### lesing av fil

vi kan lese filer via funksjonen loadtxt:
data = loadtxt('filnavn', skiprows = 1, delimiter = ',', dtype = str)

skiprows sier om du skal de bort i fra en rad når du printer

delimiter forteller hva som deler kolonnene i teks dokumentet du importerer fra

### gjennomsnitt og standardavvik

for å få gjennomsnitt bruker man funksjonen mean(...)

standardavvik er en verdi for spredning av verdiene i et datasett 
for å få standardavvik bruker man funksjonen std(...)

```python
from pylab import *

data = loadtxt('sunspots.txt', skiprows = 1, delimiter = ',')

nr = data[:,0]   #printer alle radene i kolonne 0
antall = data[:,1]

snitt = mean(antall)    #gjennomsnitt
avvik = std(antall)     #standard deviation

plot(nr, antall)
xlabel('årstall')
ylabel('antall solflekker')

print('snitt:', snitt, 'standardavvik:', avvik)
```

### regresjon

regresjon betyr å finne ut en funksjon som passer best mulig for alle pun 

for å utføre en regresjon bruker du polyfit funksjonen basert på grad

#### førstegrads regresjon

```python
from pylab import*

t = [0, 20, 40, 60, 80, 100]
los_NH3 = [88.5, 56.0, 34.0, 20.0, 11.0, 7.0]
los_NaCl = [35.7, 35.9 ,36.4, 37.1, 38.0, 39.2]



grad = 1
reg_NaCl = polyfit(t, los_NaCl, grad)

x = linspace(0, 150, 1000)

a = reg_NaCl[0]
b = reg_NaCl[1]

y = a*x + b

plot(x,y)
scatter(t, los_NaCl)   #graf uten linjer
```

her får vi svaret i formen ax + b da er svarene i arrayen a og b

#### andregradsregresjon

```python
from pylab import*

t = [0, 20, 40, 60, 80, 100]
los_NH3 = [88.5, 56.0, 34.0, 20.0, 11.0, 7.0]
los_NaCl = [35.7, 35.9 ,36.4, 37.1, 38.0, 39.2]



grad = 2
reg_NaCl = polyfit(t, los_NaCl, grad)

x = linspace(0, 150, 1000)

a = reg_NaCl[0]
b = reg_NaCl[1]
c = reg_NaCl[2]

y = a*x**2 + b*x + c

plot(x,y, color = 'red')
scatter(t, los_NaCl)   #graf uten linjer
```

her får vi en andregradsfunksjon med formen ax^2 + bx + c

## regresjonsanalyse
det finnes mange typer regresjon og de ulike metodene gir ulike kurver

vi kan teste hvor godt modellen vår passer med dataene våre ved å gjøre ulike analyser vanlig analyse er å finne r2 (determinasjonskoeffesienter) 

r2 ligger mellom 0 og 1 der en er bestmulig tillpassning

```python
from pylab import*
from sklearn.metrics import r2_score    #for å skjekke verdier

t = [0,20,40,60,80,100]
tarray = array(t)
los_NH3 = [88.5,56.0,34.0,20.0,11.0,7.0]
los_NaCl = [35.7, 35.9 ,36.4, 37.1, 38.0, 39.2]


grad = 2
reg_NH3 = polyfit(t, los_NH3, grad)

x = linspace(0, 150, 1000)

a = reg_NH3[0]
b = reg_NH3[1]
c = reg_NH3[2]

y = a*x**2 + b*x + c
y_r2 = a*tarray**2 + b*tarray + c

r2 = r2_score(los_NH3, y_r2)
print(r2)

plot(x,y, color = 'red')
scatter(t, los_NH3)   #graf uten linjer
```

## regresjon = modell
når vi utfører regresjon lager vi en modell for utviklingne i datasettene våre
med en modellkan vi ekstrapolere (forutssi framtidig utvikling)

#### en god r2 verdi betyr ikke en god graf!!!

## statiestisk analyse

en statiestisk analyse kan avdkke trender og mønstre i dataene våre
det er viktig å ha kontroll på feil og måleusikkerhet
vi gjør ofte mange målinger for å minske usikkerheten i måledataene våre



### |gjennomsnitt|
gjennomsnittet av n målinger er definert som sum av alle målingene dividert på n 

$$\bar x = \frac{1}{n} \sum_{i=1}^n x_i $$

## Målinger på spredning av dataene
vi har flere målinger for spredning av målinger: varians of standardavvik er det vanligste

standardavviket er definert som den possitive kvadratrota av varansen, og er den vanligste måten å måle statistisk spredning

$$\sigma = \sqrt{\frac{1}{n} \sum_{i=1}^n (x_i - \bar x)} $$
(std funksjon)
